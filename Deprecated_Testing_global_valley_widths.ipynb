```python
def find_nearest(array): 
    
    loc_maxs= ssig.argrelmax(array)
    glob_min= np.argmin(array)
    
    id1 = (np.abs(loc_maxs[0]-glob_min)).argmin()
    id2 = ((loc_maxs[0]-glob_min)>0).argmax()
        
    return loc_maxs[0][id1],loc_maxs[0][id2]
```

```python
x= np.array([2,1,3,4,3,2,3,0,5,6,7,5,4,3,1,8,6,8])
near_maxs= find_nearest(x) # 2 maximum indices nearest the global minimum
```

```python
print(near_maxs)

plt.plot(np.arange(len(x)),x, "b", near_maxs[0],x[near_maxs[0]],"rx", near_maxs[1],x[near_maxs[1]],"g^")
plt.plot(np.arange(near_maxs[0], near_maxs[1]+1),np.max(x)*np.ones((near_maxs[1]-near_maxs[0]+1)),"k")
plt.show()
```
